```python
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from itertools import product, combinations



fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111,projection='3d')
ax.set_aspect("equal")


# ###########
# draw cube
#############
r = [0, 1]
for s, e in combinations(np.array(list(product(r, r, r))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        ax.plot3D(*zip(s, e), color="b")

###################
# cartesian 3D axis
###################

# axis limits (numbers)
ax.set_xlim(-0.01,2)
ax.set_ylim(-0.01,2)
ax.set_zlim(-0.01,2)
ax.view_init(elev=20., azim=32)

# axis labels x, y, z
ax.text(1.95, 0, 0, 'x', ha='center')
ax.text(0, 1.95, 0, 'y', ha='center')
ax.text(0, 0, 1.95, 'z', ha='center')


# axis arrows (black)
x, y, z = np.zeros((3,3))
u, v, w = np.array([[1.8,0,0],[0,1.8,0],[0,0,1.8]])
ax.quiver(x,y,z,u,v,w,arrow_length_ratio=0.1,color='k')


plt.show()
```
